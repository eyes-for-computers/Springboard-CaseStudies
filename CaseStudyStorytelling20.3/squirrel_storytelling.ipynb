```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Every spring as all the animals in the forst being to come to life, there are more and more humans that want to 
# admire them. Since the pandemic began people have been looking for more satisfying and relaxing hobbies. 
# "Briding" is one of the fastest growing animal watching activities, but in some areas birds are in short supply. 
# There is another option for these animal lovers: Squirrels. This exercise examines the nyc_squirrels database.
# Audience: non-technical
#
# Questions to answer:
#
# What is the best day to see a squirrel? Common and rare squirrels?
# When is the best time to feed squirrels?
```

```python
s = pd.read_csv('nyc_squirrels.csv')
print('dtypes: ', + s.dtypes)
#s = s[cols_to_keep]
print('\nshape:', s.shape)
```

```python
s.zip_codes.unique()
```

```python
movements = ['running', 'chasing', 'climbing', 'tail_twitches', 'approaches', 'tail_flags', 'foraging']
```

```python
# No significat corrolations were found in the dataset

s.date = pd.to_datetime(s['date'], format='%m%d%Y')
s.corr()
```

```python
days = []
for i in s.date:
    days.append(i.day_name())
s['day_name'] = days

print('How many sightings of the same squirrel?')
int(len(s.unique_squirrel_id)) - int(len(s.unique_squirrel_id.unique()))
```

```python
print('How much are squirrels observed in movement? \n')
[print(s[i].value_counts(), '\n') for i in movements]
```

```python
# Visually inspecting corrolations of squirrel movements and find nothing significant
sns.heatmap(data=s[movements])
```

```python
# Squirrels seem to be out more on the weekend, and particulary not on on Tuesday/Thursday

sns.countplot(x="day_name", data=s).set_title(
  'Squirrels seen by day', fontdict = { 'fontsize': 20})
```

```python
# Saturday and Thursday have the highest chance of being approached by a squirrel
#

saturday = s.loc[s['day_name'] == 'Saturday']

plt.figure(figsize = (15,8))
sns.lineplot(x=s.day_name, y=s.approaches).set_title(
  'Number of approaches by day', fontdict = { 'fontsize': 20})

```

```python
# Gray is the most common squirrel type and black is rare

sns.countplot(x='primary_fur_color', data=s).set_title(
  'How common are each type of squirrel?', fontdict = { 'fontsize': 20})
```

```python
print(s.primary_fur_color.value_counts())
# You're most likely to find a rare black squirrel on Saturday

black = s.loc[s['primary_fur_color'] == 'Black']
sns.countplot(x='day_name', data=black).set_title(
  'Black squirrels by day', fontdict = { 'fontsize': 20})
```

```python
plt.figure(figsize = (15,8))

sns.lineplot(x = 'day_name', y = 'approaches', hue='primary_fur_color',data = s).set_title(
  'Number of approaches by squirrel type', fontdict = { 'fontsize': 20})
```

```python
plt.figure(figsize = (15,8))
#sns.lineplot(x = 'shift', y = 'indifferent',data = s)
sns.lineplot(x = 'day_name', y = 'approaches', hue='shift',data = s).set_title(
  'Number of approaches by day', fontdict = { 'fontsize': 20})
```

```python
# Conclusion:
# 
# If you're new to squirrel watching then go on the weekends for the best luck to see a squirrel. Tuesday and
# have the lowest squirrel sightings overall. You'll find the gray squirrels to be the most common and black to 
# be the most rare. You will the best chance to find more rare squirrels on saturday.

# If you are brave enough to want to feed the squirrels you will have the most luck with cinnamon squirrels
# Though not the most common, you are nearly twice as likely to be approached by a cinnamon squirrel than 
# gray or black. The best time is Saturday mornings. 
```
